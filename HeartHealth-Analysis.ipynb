```python
# Import library
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Dataset
df = pd.read_csv('Heart Disease data.csv')

# View information of dataset 
df.head(10)
```

```python
df.shape # Number of columns present in the dataFrame
```

```python
df.isnull() # Detect missing values
```

```python
df.isnull().sum() # Returns the number of missing values in the dataset
```

```python
df.isnull().sum().sum() # Total numbers of null values
```

```python
df = df.dropna(axis=0) # By default axis=0 , 0 for rows and 1 for columns....
```

```python
df.shape # Number of columns present in the dataFrame
```

```python
df = df.drop_duplicates() # No duplicates in dataset
```

```python
df.shape # Number of columns present in the dataFrame
```

```python
df.info() # Information about the DataFrame
```

```python
df.corr() #  find the pairwise correlation
```

## Exploratory Data Analysis `EDA`

- Now Start the `Visualization` Part

```python
sns.heatmap(df.corr(),annot=True)             # Plot heatmap with correlation values
plt.title('Correlation among all the attributes',fontsize=25)
plt.show
```

```python
sns.countplot(x="target",data=df)  #countplot to check how many people have heart disease
plt.xticks(([0,1]),['Less chance of heart attack', 'More chance of heart attack'])
plt.xlabel('Target',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.show()
```

```python
df["sex"].value_counts()     #Find count of Male and Female in this Dataset   
sns.countplot(x = "sex", data = df) 
plt.xticks(([0,1]),['Female', 'Male'])
plt.xlabel('Chest Pain',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Chest Pain by Targeted People',fontsize=25)
plt.show()
```

```python
sns.countplot(x='sex',hue='target',data=df)
plt.xticks(([1,0]),['Male','Female'])
plt.legend(labels=['No-Disease', 'Disease'])
plt.xlabel('Gender',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Gender Count based on heart disease',fontsize=25)
plt.xticks([1,0],['Male','Female'])
plt.show
```

```python
sns.countplot(data=df,x='cp',hue='target',palette='CMRmap')
plt.legend(labels=['No-Disease', 'Disease'])
plt.xticks(([0,1,2,3]),['typical angina','atypical angina', 'non-anginal pain','asymptomatic'])
plt.xlabel('Chest Pain',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Chest Pain by Targeted People',fontsize=25)
plt.show()
```

```python
sns.lineplot(x='cp',data=df,y='age',hue='sex')
plt.legend(labels=['Male','Female'])
sns.set(rc={'figure.figsize':(15,5)})
plt.xlabel('Chest Pain',fontsize=25)
plt.ylabel('Age',fontsize=25)
plt.title('Chest Pain by Age',fontsize=25)
plt.show()
```

```python
sns.countplot(data=df,x='fbs',hue='target')
plt.legend(labels=['No-Disease', 'Disease'])# fasting biood sugar distribution according
plt.xlabel('Blood Sugar',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('FBS by Targeted People',fontsize=25)
plt.show
```

```python
sns.histplot(x='trestbps',data=df,color='black')
plt.xlabel('Trestbps',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Trestbps',fontsize=25)
plt.show
```

```python
sns.countplot(x='age',data=df,hue='target')
plt.xlabel('Age',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Age count by heart disease',fontsize=30)
plt.show
```

```python
sns.histplot(x='chol',data=df,color='r') # show distribution of serum cholesterol
plt.xlabel('Cholesterol',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Cholesterol',fontsize=25)
plt.show
```
